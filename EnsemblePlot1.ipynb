```python
#import relevant spacepy modules
import spacepy.toolbox as tb
import spacepy.omni as om
import spacepy.datamanager as dmng
import spacepy.time as spt
```

```python
#import relevant python modules
import numpy as np
import matplotlib.pyplot as plt
```

First set up a time series, I'll use the hourly OMNI for demonstration purposes

```python
ticks = spt.tickrange('2007-01-01T00:00:00', '2007-01-13T23:59:59', 1/24.0)
omdata = om.get_omni(ticks)
```

Let's use the solar wind velocity as an example - in this data structure that's named 'velo'

```python
plt.plot(omdata['UTC'], omdata['velo'], 'k-')
```

Since I don't have the kernels for the binned distributions I'll assume that they're in a function called model that takes a value, determines what bin it's in and returns a random value from the appropriate distribution. For ease of illustration my models will all be identical Gaussians...

```python
def model(in_val, nsamples=1):
    mu = (np.round(in_val/50.0))*50.0 #rounds to nearest 50
    spread = 37.0
    return np.random.normal(mu, spread, nsamples)
```

I'm going to be inefficient here and just use a list comprehension to loop over all elements of the time-series and generate surrogates

```python
surr_data = np.array([model(v, nsamples=1000) for v in omdata['velo']])
```

```python
mad = tb.medAbsDev(surr_data)
lower = omdata['velo'] - mad
upper = omdata['velo'] + mad
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(omdata['UTC'], omdata['velo'], 'k-')
ax.fill_between(omdata['UTC'], lower, upper, alpha=0.5, facecolor='blue')
```

Now to add just a couple of example surrogate time series... (and pretty the plot up a bit)

```python
fig = plt.figure()
ax = fig.add_subplot(211)
ax.plot(omdata['UTC'], omdata['velo'], 'k-', lw=2)
for n in [9,201,374,669,923]:
    ax.plot(omdata['UTC'], surr_data[:,n], ':')
ax.fill_between(omdata['UTC'], lower, upper, alpha=0.3, facecolor='blue')
ax.set_ylabel('V$_{sw}$ [km/s]')
labels = ax.get_xticklabels()
dum = plt.setp(labels, rotation=45)

#and the second plot is just a zoomed version
ax2 = fig.add_subplot(212)
ax2.plot(omdata['UTC'][0:120], omdata['velo'][0:120], 'k-', lw=2)
for n in [9,201,374,669,923]:
    ax2.plot(omdata['UTC'][0:120], surr_data[0:120,n], ':')
ax2.fill_between(omdata['UTC'][0:120], lower[0:120], upper[0:120], alpha=0.3, facecolor='blue')
ax2.set_ylabel('V$_{sw}$ [km/s]')
labels = ax2.get_xticklabels()
dum = plt.setp(labels, rotation=45)
```
