```python
import numpy as np
import matplotlib.pyplot as plt
import spacepy.plot as splot
%matplotlib inline

splot.style('spacepy')
```

```python
def APE(pred, obs):
    err = pred-obs
    are = np.abs(err/obs)
    return 100*are
```

```python
x = np.logspace(-1, 8, 1000)
y = x*10.
goodape = APE(y,x)
badape = APE(np.log(y), np.log(x))
print(np.mean(goodape))
print(np.mean(badape))
```

```python
plt.semilogx(x, badape, 'r-', label='log(x), log(y)')
plt.semilogx(x, goodape, 'k-', label='x, y')
plt.ylim([0, 1250])
plt.ylabel('APE [%]')
plt.xlabel('Observed value')
plt.legend()
plt.title('APE for y=10x')
```
